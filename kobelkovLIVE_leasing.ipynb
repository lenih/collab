```python
!pip install folium pandas geopy

import folium
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut

# Данные по регионам и основным сферам лизинга
data = {
    "Регион": [
        "Москва", "Санкт-Петербург", "Краснодар", "Иркутск", "Иркутская область",
        "Новосибирск", "Казань", "Екатеринбург", "Владивосток", "Ростов-на-Дону"
    ],
    "Основная сфера": [
        "Легковые автомобили", "Спецтехника", "Сельхозтехника", "Медицинское оборудование", "Грузовые автомобили",
        "Железнодорожный транспорт", "Оборудование для пищевой промышленности", "Строительная техника", "Авиационный транспорт", "Энергетическое оборудование"
    ]
}

df = pd.DataFrame(data)

# Определение цветов для разных сфер
category_colors = {
    "Легковые автомобили": "red",
    "Спецтехника": "blue",
    "Сельхозтехника": "green",
    "Медицинское оборудование": "purple",
    "Грузовые автомобили": "orange",
    "Железнодорожный транспорт": "cadetblue",
    "Оборудование для пищевой промышленности": "darkgreen",
    "Строительная техника": "pink",
    "Авиационный транспорт": "lightblue",
    "Энергетическое оборудование": "darkred"
}

# Создание карты
m = folium.Map(location=[61.5240, 105.3188], zoom_start=4)

# Геокодирование (определение координат по названию региона)
geolocator = Nominatim(user_agent="geoapiExercises")

for _, row in df.iterrows():
    try:
        location = geolocator.geocode(row["Регион"])
        if location:
            color = category_colors.get(row["Основная сфера"], "gray")  # Если нет цвета, используется серый
            folium.Marker(
                [location.latitude, location.longitude],
                popup=f"<b>{row['Регион']}</b><br>Сфера: {row['Основная сфера']}",
                icon=folium.Icon(color=color)
            ).add_to(m)
    except GeocoderTimedOut:
        print(f"Ошибка геокодирования для региона {row['Регион']}")

# Сохранение карты
m.save("map.html")
m
```
